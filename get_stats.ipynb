```python
import numpy as np
import pandas as pd
```

```python
# Load in pitcher and hitter data for the whole league
pitchers = pd.read_csv('pitchers.txt')
batters = pd.read_csv('batters.txt')

# Read in team data
astros = pd.read_csv('astros.txt')

def clean_team_data(team):
    # Remove the first row of astros and third column
    team = team.drop(team.index[0])
    team = team.drop(team.columns[2], axis=1)

    # Rename the 4th column to Home/Away
    team = team.rename(columns={'Opp' : 'Opponent',
                                'Tm': 'Team',
                                'Unnamed: 4': 'Home/Away'})

    # Change the Home/Away column to 1 for home and 0 for away
    team['Home/Away'] = team['Home/Away'].map({'@': 'Away', '': 'Home'})

    return team
```

```python
astros = clean_team_data(astros)
astros.head()
```
